### - Exercici 1

Descarrega el data set Airlines Delay: Airline on-time statistics and delay causes i carrega’l a un pandas Dataframe. Explora les dades que conté, i queda’t únicament amb les columnes que consideris rellevants.

```python
import pandas as pd
import numpy as np


csv = pd.read_csv(r'C:\Users\motxi\Documents\DelayedFlights.csv')
df = pd.DataFrame(csv)
print(df.head(), '\n')
print(df.info())  #mirem les columnes amb la info que ofereixen
```

```python
#Inspecciono les columnes que crec mes rellevants, a modo de taules les agrupo en 3 grans categories ; 

#-Temporals: *(Relatives al calendari i temps de vol) 

df.loc[:, ['Year','Month', 'DayofMonth', 'DayOfWeek', 'AirTime','CarrierDelay', 'ArrDelay', 'DepDelay']].head()
```

```python
#-Causes: *(Relatives a les causes que orginen el retard)

df.loc[:, ['WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']].head()
```

```python
#-Vol: *(Miscelanea d'objectes relacionats amb el trajecte i l'aerolinia )

df.loc[:, ['UniqueCarrier', 'Origin', 'Dest', 'Distance', 'Diverted', 'Cancelled', 'CancellationCode']].head()
```

```python
#Filtro totes les columnes que no vull per crear les noves taules. 

#Creo la taula temporals
temporals = df.filter(['Year','Month', 'DayofMonth', 'DayOfWeek', 'AirTime', 'ArrDelay', 'DepDelay'], axis=1)

#Creo la taula causes
causes = df.filter(['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay'], axis=1)

#Creo la taula vol
vol = df.filter(['UniqueCarrier', 'Origin', 'Dest', 'Distance', 'Diverted', 'Cancelled', 'CancellationCode'], axis=1)

#Creo la unió de les tres taules anteriors

data = pd.concat([temporals, causes, vol],axis=1)
data.head()
```
